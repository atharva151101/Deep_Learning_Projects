```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df0=pd.read_csv("../../20/csv/elefante.csv",header=None)
df1=pd.read_csv("../../20/csv/farfalla.csv",header=None)
df2=pd.read_csv("../../20/csv/gatto.csv",header=None)
df3=pd.read_csv("../../20/csv/cavallo.csv",header=None)
df4=pd.read_csv("../../20/csv/ragno.csv",header=None)
```

```python
cols={}
for i in range(48):
    cols[i+1]=f'x{i}'

df0=df0.loc[:,1:48]
df1=df1.loc[:,1:48]
df2=df2.loc[:,1:48]
df3=df3.loc[:,1:48]
df4=df4.loc[:,1:48]

df0=df0.rename(columns=cols)
df1=df1.rename(columns=cols)
df2=df2.rename(columns=cols)
df3=df3.rename(columns=cols)
df4=df4.rename(columns=cols)

df0['y']=0
df1['y']=1
df2['y']=2
df3['y']=3
df4['y']=4
```

```python
df=pd.concat([df0,df1,df2,df3,df4],axis=0)
df
```

```python
a=df.to_numpy()
a=a.reshape(a.shape[0]*a.shape[1])
a.shape
```

```python
plt.hist(a,bins=100)
plt.show()
```
